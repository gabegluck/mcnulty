###Defining Functions

```python
def process_edu_level_name(edu):
    """Combine all under high school,
    """
    if edu == "High school diploma or equivalent":
        return "High School Graduate"
    elif edu == "Bachelor's degree":
        return edu
    elif edu == "Some college but no degree":
        return "Some college"
    elif edu == "Associates degree - academic program" or edu == "Associates degree - occupational/vocational program":
        return "Associates degree"
    elif edu == "Doctorate degree" or edu == "Master's degree" or edu == "Professional school degree":
        return "Graduate Degree"
    else:
        return "Did not complete high school"
    
def process_race_name(race):
    """ Keep Asian, White, African American, combine all others since they're less than .01 of dataset into other
    """
    race_list=['White', 'Asian only', 'Black/Negro']
    if race == 'White':
        return race
    elif race =='Asian only':
        return 'Asian'
    elif race == 'Black/Negro':
        return "African American"
    else:
        return "Other Race"
def process_marital_status(marital):
    """ Combine both married groups as well as  separated and divorced
    """
    if marital== "Married spouse present" or marital == "Married spouse absent":
        return "Married"
    elif marital == "Never married/single":
        return "Single"
    elif marital == "Divorced" or marital == "Separated":
        return "Separated/Divorced"
    else:
        return marital
def process_columns(df,columns):
    """apply processing
    """
    
    dg=df.copy()
    for i in columns:
        dg[i]=dg[i].apply(eval("process_"+str(i)))
    return dg

#making proportion of each column, for clustering
def get_proportion_of_column(df,column):
    """ input: dataframe, column in data frame with ordinal categories 
    i.e gender, df.gender.unique() -> array(['Female', 'Male'], dtype=object)
    output: column grouped by occupation for proportion of each value in the input column, unique values for column
    """
    
    #get unique column values
    unique_values=df[column].unique()
    
    #get dataframe with all counts
    grouped_df=df.groupby(['occupation_name',column]).count().age
    #get length of uniques
    len_unique=(len(list(unique_values)))
    proportion_whole_column=[]
    occupation_var=list(df.occupation_name.unique())
    for occ in occupation_var:
        counts=[]
        #look at each value to see if it exists
        for i in unique_values:
            try:
                counts.append(grouped_df[occ][i])
            except:
                counts.append(0)
        prop_occ=[float(i)/sum(counts) for i in counts]
        proportion_whole_column.append(prop_occ)
    return proportion_whole_column, unique_values

def make_dataframe(proportions_list,column_names):   
    """input: list of proportion and list of value_names
    output: dataframe
    """
    occupation_var=list(df.occupation_name.unique())
    proportiondf=pd.DataFrame(data=proportions_list, index=occupation_var, columns=column_names)
    return proportiondf

def make_concatanated_dataframe(df,columns):
    list_of_dataframes=[]
    occupation_var=list(df.occupation_name.unique())
    for column in columns:
        if df[column].dtype == np.int64:
            mean_col=df.groupby('occupation_name').mean()[column]
            list_of_dataframes.append(mean_col)
        else:
            proportion_each_column, column_names=get_proportion_of_column(df, column)
            mean_data_frame=make_dataframe(proportion_each_column,column_names)
            list_of_dataframes.append(mean_data_frame)
    bigdf=pd.concat(list_of_dataframes, axis=1)
    
    return bigdf

  
```

###Do Preprocessing

```python
columns_to_change=['edu_level_name',"race_name", 'marital_status']
processed_cps_df=process_columns(df,columns_to_change)
#make X and Y data
Xvars=processed_cps_df[['age', 'edu_level_name','race_name', 'gender']]
Yvars=processed_cps_df.occupation_name
#make dummies for edu,marital, race, gender
initial_dummies=[u'gender', u'race_name', 'edu_level_name']
for i in initial_dummies:
    series=pd.get_dummies(Xvars[i])
    Xvars=Xvars.join(series)
    Xvars=Xvars.drop(i, axis=1)

    
```

```python
demographic_var=['age','gender', 'race_name', 'edu_level_name'] 
concatanated=make_concatanated_dataframe(processed_cps_df,demographic_var)
concatanated_scaled=preprocessing.scale(concatanated)
concatanated.head()
```

###Make clusters

```python
concatanated_pca = decomposition.PCA(n_components=5).fit_transform(concatanated_scaled)
clustering=cluster.AgglomerativeClustering(n_clusters=6)
clustering.fit(concatanated_pca)
cluster_ids= clustering.fit_predict(concatanated_pca)

clusters = defaultdict(list)
for clu, occup in zip(cluster_ids, list(concatanated.index)):
    clusters[clu].append(occup)
    
for clu, occupations in clusters.iteritems():
    print "_______%i_______" % clu
    for occup in occupations:
        print occup
    print
    print
```

```python
#assign cluster values to occupationss
Y_clusters=[]
for i in Yvars:
    for key in clusters.keys():
        for value in clusters[key]:
            if i==value:
                Y_clusters.append(key)
```

### Run Models

```python
Xtrain, Xtest, Ytrain, Ytest=train_test_split(Xvars, Y_clusters)
Xtrain=preprocessing.scale(Xtrain)
Xtest=preprocessing.scale(Xtest)

#Logistic Regression
logreg =  linear_model.SGDClassifier(loss='log')
logreg.fit(Xtrain,Ytrain)
svmmodel=linear_model.SGDClassifier()
svmmodel.fit(Xtrain,Ytrain)

#Gaussian Naive Bayes 
nb=GaussianNB()
nb.fit(Xtrain,Ytrain)
print "Gaussian Naive Bayes Model Accuracy: %.5f "% accuracy_score(Ytest, nb.predict(Xtest))

#bernoulli nb
bnb=BernoulliNB()
bnb.fit(Xtrain,Ytrain)
print "bernoulli nb Accuracy: %.5f"% accuracy_score(Ytest, bnb.predict(Xtest))

#Decision Tree

tree=DecisionTreeClassifier()
tree.fit(Xtrain,Ytrain)
print "DecisionTreeClassifier Accuracy: %.5f"% accuracy_score(Ytest, tree.predict(Xtest))

#Random Forest

rf=RandomForestClassifier()
rf.fit(Xtrain,Ytrain)
print "RandomForestClassifier Accuracy: %.5f"% accuracy_score(Ytest, rf.predict(Xtest))


print "Logisitic Regression Accuracy: %.5f"% accuracy_score(Ytest, logreg.predict(Xtest))
print "SVM Accuracy: %.5f"% accuracy_score(Ytest, svmmodel.predict(Xtest))

#look at some classification reports
print 'Bernoulli Naive Bayes'
print classification_report(Ytest, bnb.predict(Xtest))
print 'Logistic Regression'
print classification_report(Ytest, logreg.predict(Xtest))


    
```
